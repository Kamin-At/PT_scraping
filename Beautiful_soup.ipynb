```python
from bs4 import BeautifulSoup as soup  # HTML data structure
from urllib.request import urlopen as uReq  # Web client
```

```python
a = 'เฟิสช้อย'.encode(encoding='utf8')
# page_url = f"https://pantip.com/search?q={a}"
page_url = "https://pantip.com/topic/35281083"
```

```python
uClient = uReq(page_url)

# parses html into a soup data structure to traverse html
# as if it were a json data type.
page_soup = soup(uClient.read(), "html.parser")
uClient.close()

# finds each product from the store page
containers = page_soup.findAll("div", {"class": ["desc col-md-12"]})#, "display-post-story-wrapper comment-wrapper"]})
```
